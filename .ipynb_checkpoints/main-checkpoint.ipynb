```python
#!/usr/bin/env python  
# -*- coding: utf-8 -*-

""" 
@version: v1.0 
@author: SHIFengCN 
@license: MIT 
@contact: shifree@gmail.com
@site: https://github.com/shifengcn 
@software: PyCharm 
@file: main1.py 
@time: 2019/6/26 0026 18:59 
"""

import pyttsx3
import pandas as pd
import re
import xlrd
df=pd.read_excel("chinese.xlsx",header=None,)
df
```

```python
import itertools as it
#import math
engine = pyttsx3.init()
#engine.say("总共有"+str(df.shape[0])+"课")
engine.runAndWait()
is_exit=False
while not is_exit :
    #engine.say("按课默写请输入一回车，按年级默写请输入二回车")
    #engine.runAndWait()
    choice1=input("按课默写请输入1,回车，按年级默写请输入2回车，返回请输入back，退出请输入exit：")
    if choice1=='back' or choice1=='b':
        break
    elif choice1=='exit' or choice1=='e':
        is_exit=True
        break
    elif choice1=='1':
        while  not is_exit:
            grade1=input('请输入年级，请使用阿拉伯数字，返回请输入back，退出请输入exit：')
            if grade1=='back' or grade1=='b':
                break
            elif grade1=='exit' or grade1=='x':
                is_exit=True
                break
            elif int(grade1) in df.iloc[:,0].values.tolist():
                while not is_exit:
                    grade2=input('请输入上下册，上册输a，下册输b，请使用小写字母，返回请输入back，退出请输入exit：')
                    if grade2=='back' :
                        break
                    elif grade2=='exit' or grade2=='x':
                        is_exit=True
                        break
                    elif grade2=='a' or grade2=='b':
                        while not is_exit:
                            #print(set(df[df[0]==2][df[1]=='a'].iloc[:,2].values.tolist()))
                            unit1=input('请输入课号，请使用阿拉伯数字，返回请输入back，退出请输入exit：')
                            #print()
                            if unit1=='back' or unit1=='b':
                                break
                            elif unit1=='exit' or unit1=='x':
                                is_exit=True
                                break
                            elif int(unit1) in df[(df[0]==int(grade1))&(df[1]==grade2)].iloc[:,2].values.tolist():
                                word_line=set(df[(df[0]==int(grade1))&( df[1]==grade2)&( df[2]==int(unit1))].iloc[0,3:].values.tolist())
                                #print(word_line)
                                for word in word_line:
                                    is_goback=False
                                    if is_exit:
                                        break
                                    # nan类型是float，一般如果判断使用math.isnan(float(word))
                                    #print(type(word))
                                    if (word is not None) and not isinstance(word,float):
                                        #print(word)
                                        while not is_exit:
                                            engine.say(word)
                                            engine.runAndWait()
                                            command = input("请按回车切换下一个单词，按r重听，按b返回，按x退出：")
                                            if command=='r':
                                                continue
                                            elif command=='x' or command=='exit':
                                                is_exit=True
                                            elif command=='b' or command=='back':
                                                is_goback=True
                                            break
                                    if is_goback:
                                        break
                                print('播放完毕')            
                            else:
                                print("无此课号，请重新输入！")
                    else:
                        print('输入错误，请重新输入！')
            else:
                print('输入错误，请重新输入!')
    elif choice1=='2':           
        while not is_exit:
            grade1=input('请输入年级，请使用阿拉伯数字，返回请输入back，退出请输入exit：')
            if grade1=='back' or grade1=='b':
                break
            elif grade1=='exit' or grade1=='x':
                is_exit=True
                break
            elif int(grade1) in df.iloc[:,0].values.tolist():                
                while not is_exit:
                    grade2=input('请输入上下册，上册输a，下册输b，请使用小写字母，返回请输入back，退出请输入exit：')
                    if grade2=='back' :
                        break
                    elif grade2=='exit' or grade2=='x':
                        is_exit=True
                        break
                    elif grade2=='a' or grade2=='b':
                        while not is_exit:
                            is_goback=False
                            word_lines=set(it.chain.from_iterable(df[(df[0]==int(grade1))&(df[1]==grade2)].iloc[:,3:].values.tolist()))
                            #print(word_lines)
                            for word in word_lines:
                                if is_exit:
                                    break
                                if (word is not None) and not isinstance(word,float):
                                    #print(word)
                                    while not is_exit:
                                        engine.say(word)
                                        engine.runAndWait()
                                        command = input("请按回车切换下一个单词，按r重听，按b返回，按x退出：")
                                        if command=='r':
                                            continue
                                        elif command=='x' or command=='exit':
                                            is_exit=True
                                        elif command=='b' or command=='back':
                                            is_goback=True
                                        break
                                if is_goback:
                                    break
                            if is_goback:
                                break
                    else:
                        print('无此年级单词，请重新输入！')                                
            else:
                print('输入错误，请重新输入!')    
    else:
        print('输入错误，请重新输入！')
```
