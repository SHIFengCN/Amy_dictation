```python
import pandas as pd
import time
import ybc_speech
df=pd.read_excel("chinese.xlsx",header=None,)
line_count=df.shape[0]
text_list=list()
word_list=list()
j=0
for line in range(line_count):
    #print(df.iloc[line,3:].tolist())
    text=[i for i in df.iloc[line,3:].tolist() if (not isinstance(i,float)) ]
    #print('_'.join(text))
    #text_list.append('_  ,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,   _'.join(text))
    #print(text)
    for tex in text:
        file=ybc_speech.text2voice(str(j)+',,,,,,,,,,,,,,,,,,'+tex, 'd:\\downloads\\语文默写\\'+str(j)+'.wav', speaker=1, speed=0.5, aht=0, apc=58, volume=100,
            _format=2)
        word_list.append(str(j)+'_'+tex)
        print(tex)
        print(file)
        #time.sleep(3)
        j=j+1
#print(text_list)
print(word_list)


```
